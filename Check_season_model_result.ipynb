```python
team = {}
for gw in range(1, 39):
    team[gw] = {
        "line_up" : {},
        "bench" : [],
        "total_points" : 0
    }
    
variables = {
    "player_on_team" : [],
    "player_on_lineup" : [],
    "player_is_captain" :[]
}



with open("models/result.log", "r") as file:
    ready = False
    while not ready:
        text = file.readline()
        ready = "objective value:" in text
    objective_value = int(text.split(" ")[-1])
    
    ready = False
    while not ready:
        text = file.readline()
        if text == "\n":
            break
            
        variable, player = text.split("$")
        player, gw = player.split("#")
        gw = gw.split(" ")
        gw, obj = (int(gw[0]), gw[-1])
        obj = int(obj.split(":")[1].replace(")", ""))
        
        
        if variable == "player_on_team":
            team[gw]["bench"].append(player)
        elif variable == "player_on_lineup":
            team[gw]["bench"].remove(player)
            team[gw]["line_up"][player] = obj
            team[gw]["total_points"] += obj
        elif variable == "player_is_captain":
            team[gw]["line_up"].pop(player)
            team[gw]["line_up"][f"{player} (C)"] = obj
            team[gw]["total_points"] += obj
        elif variable == "player_is_transfered":
            print("Existe!")
            break
        
        variables[variable].append(f"{player}|{gw}")

team    
```

```python
variables["player_is_transfered"] = {}
for var in variables["player_on_team"]:
    player, gw = var.split("|")
    gw = int(gw)
    
    # El jugador esta en el equipo 
    # veo si fue adquirido en esta jornada.
    var_bought = f"{player}|{gw-1}"
    if var_bought in variables["player_on_team"]:
        variables["player_is_transfered"][f"{var_bought}"] = 0
    else:
        variables["player_is_transfered"][f"{var_bought}"] = -1
        print(player, "was bought on gameweek", gw)
    
    # veo si fue vendido en la jornada siguiente
    
```
