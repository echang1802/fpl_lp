```python
team = {}
for gw in range(1, 39):
    team[gw] = {
        "line_up" : {},
        "bench" : [],
        "total_points" : 0,
        "transfers" : {
            "total" : 0,
            "players_sold" : [],
            "players_bought" : []
        }
    }
    
variables = {
    "player_on_team" : [],
    "player_on_lineup" : [],
    "player_is_captain" :[],
    "player_is_sold" : [],
    "player_is_transfered" : []
}



with open("models/result.log", "r") as file:
    ready = False
    while not ready:
        text = file.readline()
        ready = "objective value:" in text
    objective_value = int(text.split(" ")[-1])
    
    ready = False
    while not ready:
        text = file.readline()
        if text == "\n":
            break
            
        variable, player = text.split("$")
        player, gw = player.split("#")
        gw = gw.split(" ")
        gw, value, obj = (int(gw[0]), round(float(gw[-2])), gw[-1])
        obj = int(obj.split(":")[1].replace(")", ""))
        
        
        if variable == "player_on_team":
            team[gw]["bench"].append(player)
        elif variable == "player_on_lineup":
            team[gw]["bench"].remove(player)
            team[gw]["line_up"][player] = obj
            team[gw]["total_points"] += obj
        elif variable == "player_is_captain":
            team[gw]["line_up"].pop(player)
            team[gw]["line_up"][f"{player} (C)"] = obj
            team[gw]["total_points"] += obj
        elif variable == "player_is_sold":
            pass
        elif variable == "player_is_transfered":            
            if value == -1:
                team[gw]["transfers"]["players_sold"].append(player)
                team[gw]["transfers"]["total"] += 1
            elif value == 1:
                team[gw]["transfers"]["players_bought"].append(player)
            
        variables[variable].append(f"{player}|{gw}")

team    
```
